```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

start_time = time.time()

# function to get URL    
def getUrl(doc):
    if doc:
        #         CODE TO MAKE THE DICTIONARY
        localUrls = []
        # localDictionary = {'localUrls': localUrls}
        allLinks = doc.find_all('a')
        for link in allLinks:
            localUrls.append("https://commons.wikimedia.org" + link.get('href'))
        return localUrls
    else:
        return

#function to get text data from a given URL
def getTextData(url):
    response = requests.get(url)
    pageContent = response.text
    doc = BeautifulSoup(pageContent, 'html.parser')
    if doc.find(id="mw-subcategories"):
        catergoryDoc = doc.find(id="mw-subcategories")
        return catergoryDoc
    else:
        return


#function to get title data from a given URL
def getTitle(doc):
    if doc:
        #         CODE TO MAKE THE DICTIONARY
        localTitles = []
    # localDictionary = {'localTitles': localTitles}
        allLinks = doc.find_all('a')
        for link in allLinks:
            localTitles.append(link.get('title').split(':')[1])
        return localTitles
    else:
        return

#function to make a dataframe
def makedf(dict):
    df = pd.DataFrame(dict)
    return df    


site_id = "https://commons.wikimedia.org/wiki/Category:Art_of_India"

categoryTitles = []
categoryUrls = []

categoryTitles = getTitle(getTextData(site_id))
categoryUrls = getUrl(getTextData(site_id))
catergoryDictionary = {'categoryTitles': categoryTitles, 'categoryUrls': categoryUrls}
category_df = makedf(catergoryDictionary)

category_df
```

```python
subcategoryTitles = []
subcategoryUrls = []
categoryTitles1 = []
categoryUrls1 = []
start_time = time.time()
for i in category_df.categoryUrls[:1]:
    for j in range(0, len(category_df.categoryUrls[:])):
        url = getUrl(getTextData(category_df.categoryUrls[j]))
        if url:
            n = len(url)
            for k in range(0,n):
                a = category_df.categoryTitles[j]
                categoryTitles1.append(a)
                b = category_df.categoryUrls[j]
                categoryUrls1.append(b)
                c = (getTitle(getTextData(category_df.categoryUrls[j])))[k]
                subcategoryTitles.append(c)
                d= getUrl(getTextData(category_df.categoryUrls[j]))[k]
                subcategoryUrls.append(d)
                k += 1
            j += 1

subcatergoryDictionary = {'categoryTitles': categoryTitles1, 'categoryUrls': categoryUrls1, 'subcategoryTitles': subcategoryTitles, 'subcategoryUrls': subcategoryUrls}
fin_df = makedf(subcatergoryDictionary)
print(fin_df)
print("--- %s seconds ---" % (time.time() - start_time))
```
